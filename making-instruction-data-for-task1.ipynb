```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/test-data-1-10k/Test_data_1_10k.csv") 
df.head()
```

```python
metadata = pd.read_excel("/kaggle/input/adobe-test/behaviour_simulation_test_company.xlsx") #Path of xlsx file which is the original test data.
metadata.head()
```

```python
metadata = metadata[['id','date','content','username','inferred company']]
```

```python
metadata['caption'] = df['caption']
```

```python
def create_prompt(row):
    
    prompt = f"Following is the information about Twitter post."

    data_description = (f"Caption for Image of post: {row['caption']}, " 
                       f"Text content: {row['content']}, " 
                       f"Inferred company: {row['inferred company']}, " 
                       f"Username: {row['username']}, " 
                       f"Date and time: {row['date']}, " 
                   )

    prompt += data_description
    return prompt


# Apply the function to each row to create prompts
final_task1_company = pd.DataFrame()
final_task1_company['text'] = metadata.apply(create_prompt, axis=1)
final_task1_company['id'] = metadata['id']
final_task1_company.head()
```

```python
final_task1_company.to_csv("/kaggle/working/final_task1_company.csv")
```
